```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

from hnetwork import HNetwork
from visualisation import draw_bitmap
```

```python
X = np.genfromtxt('./data/large-25x25.csv', delimiter=',')
M, N = X.shape
shape=(25, 25)
```

```python
hn = HNetwork(N, train_rule='oja', update_mode='asynchronous')
```

```python
hn.fit(X, lr=1e-4, epochs=50)
```

```python
plt.imshow(hn.weights)
```

```python
plt.imshow(hn.weights)
```

```python
img_idx = 1
```

```python
image = draw_bitmap(X[img_idx,:].reshape(shape))
image.resize((100, 100), resample=Image.NEAREST)
```

```python
noise = (np.random.random(size=(1, X.shape[1])) > 0.1) * 2 - 1
x_noisy = X[img_idx, :] * noise
image = draw_bitmap(x_noisy.reshape(shape))
image.resize((100, 100), resample=Image.NEAREST)
```

```python
predicted = hn.predict(x_noisy, epochs=10)
image = draw_bitmap(predicted.reshape(shape))
image.resize((100, 100), resample=Image.NEAREST)
```
